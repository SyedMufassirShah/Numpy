# Broadcasting

In Simple, Broadcasting is basically NumPy’s way of making two different-sized arrays "match" so it can do math on them without you having to write a loop. 

Imagine you have a 3x3x3 cube and a 1x3 row. To add them, NumPy "stretches" (broadcasts) that single row to fill the entire cube. 

### The "Right-to-Left" Rule
NumPy compares the shapes of your arrays starting from the right side (the last dimension) and moving left. For each dimension, the sizes must either: 

    - Be equal.
    - One of them must be 1. 

### Visual Example
Let's look at your offsets challenge:
    - Array A (Cube): (3, 3, 3)
    - Array B (Row): ( 3) ---> NumPy treats this as (1, 1, 3) 

### How NumPy "Stretches" it:
    - Step 1: It sees the 3 at the end matches the 3 in the cube's columns.
    - Step 2: It "copies" the row [10, -5, 2] to fill all 3 rows of a layer.
    - Step 3: It "copies" that whole layer to fill all 3 depths of the cube. 

 
Now you have two 3x3x3 cubes, and NumPy just adds them element-by-element. 



```python
import numpy as np 

# Initializing a 3x3x3 array with rach index set to zero
cube = np.zeros(shape=(3,3,3), dtype=int)

# Initializing an 1x3 array 
# Which will be stretched later on durinng broadcasting to match the size of the cube 
offset = np.array([10,20,30])

# Initail Cube 
print(f"Initially: \n\n{cube}")

# Broadcasting
cube += offset
# result =  cube + offset

# After Broadcasting
print(f"\nAfter Broadcasting: \n\n{cube}")
# print(f"\nAfter Broadcasting: \n\n{result}")


```

## The "Shape-Shifter" Finale

Now that you can build, broadcast, and slice, let's try one last trick: Flattening and Transposing.

> The Challenge:
- Your cube is currently (3, 3, 3). Let's say you want to turn that 3D cube into one long 1D line of numbers.

>The Goal:
- Use a single NumPy method to turn your 27-element cube into a flat array and Print the result.

> Hint: 
- Look for a method that sounds like "making something level" or "flat" (e.g., .flatten() or .ravel()).


```python
flattened_cube = cube.flatten()
print(flattened_cube)

# Their is another method --> ravel()   which can be used to convert the 3D array into a 1D array
raveled_cube = cube.ravel()
print(raveled_cube)
```

> Spot on! You’ve got the output and the secondary method (.ravel()) down perfectly. Both do the job, though there's a tiny "under the hood" difference: 
* .flatten() always creates a copy  
* .ravel() tries to give you a view (which is faster and saves memory).


## The "Real World" Graduation Challenge

In Data Science, 3D arrays often represent Images.
> Dimension 0: Height
> Dimension 1: Width
> Dimension 2: Color Channels (Red, Green, Blue)

The Final Task:
* Imagine you have a 2x2 pixel image where each pixel has 3 color values (R, G, B).
* Create a 3D array of shape (2, 2, 3) filled with zeros.
* Using Broadcasting, set all "Red" values (the first color channel at index 0) to 255.
* Using Slicing, change the bottom-right pixel's "Green" value (index 1) to 128.

Hint:
* To target all pixels' Red channel: image[:, :, 0] = ...
* To target one specific pixel: image[row, col, channel] = ...


```python
# Setting All to Red Using Broadcasting
img = np.zeros(shape=(2,2,3), dtype=int)
red = [255,0,0]
img +=  red
print(img)
```

```python
# Using Slicing changing the bottom-right pixel to Green (128)
img[1,1,1] = 128
print(img)
```
