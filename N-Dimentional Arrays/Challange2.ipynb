```python
import numpy as np


# Task 1

# Create a 3D array named image with a shape of (3, 3, 3).
# Initialize it so that:
# The first "layer" (index 0) is all zeros (representing no Red).
# The second "layer" (index 1) is all ones (representing full Green).
# The third "layer" (index 2) is filled with the number 255 (representing full Blue).
# Bonus: Once created, use a NumPy attribute to confirm the total number of elements in the array is 27.
# How would you write the code to fill that third layer with 255 without typing the number 9 times?


# Set Layer 1 to Zeros
image = np.zeros((3,3,3), dtype=int)
# Set  Layer 2 to 1's
image[1] = 1
# Set the Layer 3 to 255's
image[2] = 255
print(image)
print(f"\n\nTotal Elements: {image.size}")
print(f"Total Elements in 3rd Layer: {image[2].size}")
```

```python
# Task 2

# Take the 3D array you just created (the 3x3x3 one)  
# Change the very center value of the middle layer (the layer of 1s) to the number 99.


image[1,1,1] = 99
print(image)
```

```python
# Task 3

# Change the first value (top-left) of every layer to the number 7.
image[:,0,0] = 7
print(image)
```

```python
# Task 4:
# In Layer 0 (the layer of zeros), change every other number in the middle row to the number 88.
# Specifically, you want Layer 0 to look like this:
# [[0, 0, 0], [88, 0, 88], [0, 0, 0]]


image[0,1,::2]=88
print(image)
```

```python
# Task 5:

flipped_image = image[::-1, :,:]
print(flipped_image)
```

```python
# Task 6


# Take your 3D image (which has 27 total elements) and transform it into a 2D matrix.

# Use the .reshape() method on your image array.
# Change it into a 2D shape that has 9 rows and 3 columns.
# Save it to a new variable called matrix_2d

reshaped_image = image.reshape(9,3)
print(f"Below is how my Reshaped Image Looks Like:\n {reshaped_image}")
print(f"\n\nShape of Reshaped Array: {reshaped_image.shape}")
print(f"Size of Reshaped Array: {reshaped_image.size}")
print(f"Dimention of Reshaped Array: {reshaped_image.ndim}")
```

```python
hotspots = np.zeros((3,3,3), dtype=bool)
print(hotspots)
print(image)

hotspots = image > 100
print(hotspots)
```

```python
# Task Only Fetch those values from image whose Mask (Reference Value) is marked True

values_found = image[hotspots]
print(values_found)
```

```python
# This finds the coordinates of every True value in your mask
indexes = np.where(image > 100)
print(indexes)
```
