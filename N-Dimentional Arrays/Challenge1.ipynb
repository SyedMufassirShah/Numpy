```python
import numpy as np

arr5x5 = np.arange(1,26).reshape(5,5)

middle_square_3x3 = arr5x5[1:4, 1:4]

print("Below is the Output 3x3 Grid:")
print(middle_square_3x3)
```

```python
# Extracting the top-left 2x2 grid 
top_left_2x2_grid = arr5x5[:2, :2] 
print("Below is the Output of Top Left 2x2 Grid:")
print(top_left_2x2_grid)
```

```python
# Grab the 1st, 3rd, and 5th rows (index 0, 2, and 4) and all the columns.
step2_jump_grid = arr5x5[::2, :]
print("Below is the Output of only 1st, 3rd and 5th Row:")
print(step2_jump_grid)
```

```python
#  Can you extract the four corners of your 5x5 grid into a new 2x2 array
corner_values = arr5x5[::4, ::4]
print(corner_values)
```

```python
#Border Values
myArr_5x5 = np.arange(1,26).reshape(5,5)

mask = np.ones((5,5), dtype=bool)

mask[1:4, 1:4] = False

border_values = myArr_5x5[mask]
print(border_values)
```

```python
# Hollow Grid
 
hollow_grid = arr5x5.copy()
hollow_grid[1:4,1:4] = 0
print(hollow_grid)
 
```
