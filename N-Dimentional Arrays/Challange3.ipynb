### Here We will Learn about Arithematic Operation in NumPy Arrays

```python
import numpy as np

array = np.random.rand(3,3,3)
print(f"Below is how Array Looks Like:\n\n{array}")
```

```python
# Sum up All the values of the Array
total_sum = np.sum(array)
print(f"Sum of all of the values of the Array is: {total_sum}")
```

```python
# Sum up only values present in the level 1:

total_sum_level_1 = np.sum(array[1,:,:])
print(f"Sum of all values of level 1 in the Array is: {total_sum_level_1}")
```

```python
# Sum of the last column of level 2 and 3
last_column_sum_level_2and3 = np.sum(array[1::, ::, -1])
print(f"Sum of last colum values of level 2 and 3 in the Array is: {last_column_sum_level_2and3}")
```

```python
# Calculate the sum of the four corners of the array
corners_sum  = np.sum([array[0,0,0],array[0,0,-1],array[2,2,0],array[2,2,-1]])
print(f"Sum of corner values of the Array is: {corners_sum}")
```

```python
# Calculate the sum of the four corners of the array (Middle Layer only)
corner_sum_level1 = np.sum([array[1,0,0],array[1,0,-1], array[1,-1,0], array[1,-1,-1]])
# corner_sum_level1 = np.sum([array[1,::2,::2]]) ----> will only work of 3x3 array 
print(f"Sum of corner values of level 1 of the array is: {corner_sum_level1}")
```

```python
# The Task:
#   Calculate the total sum of Level 1.
#   Subtract the very center value.


total_sum_level_1 = np.sum(array[1,:,:])
center_value = array[1,1,1]
print(f"Difference Between total sum of level 1 and center value: \n{total_sum_level_1-center_value}")
```

```python
# The Task:
#    Flip your 10x10x10 big_array so that:
#    The layers are reversed.
#    The rows are reversed.
#    The columns are reversed.
#    Basically, the value at [0,0,0] should end up at [9,9,9].

reversed_array = big_array[::-1,::-1,::-1]
print(reversed_array)

if big_array[9,9,9] == reversed_array[0,0,0]:
    print("True")
```
